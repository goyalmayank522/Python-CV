# GRID DETECTION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### On Chess Image

```python
flat_chess=cv2.imread('flat_chess.jpg')
plt.imshow(flat_chess)
```

```python
found, corners= cv2.findChessboardCorners(flat_chess,(7,7))
if found:
    cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
    plt.imshow(flat_chess)
else:
    print('Grid not found')
```

### On Dot Image

```python
dot=cv2.imread('dot.jpg')
plt.imshow(dot)
```

```python
found, corners=cv2.findCirclesGrid(dot,(4,4),cv2.CALIB_CB_SYMMETRIC_GRID)
if found:
    cv2.drawChessboardCorners(dot,(4,4),corners,found)
    plt.imshow(dot)
else:
    print('Grid not found')
```
