```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
reeses=cv2.imread('Data/reeses_puffs.png',0)
```

```python
display(reeses)
```

```python
cereals=cv2.imread('Data/cereals.jpg',0)
```

```python
display(cereals)
```

## Using Brute-Force ORB Descriptor

```python
orb=cv2.ORB_create()
kp1,des1=orb.detectAndCompute(reeses,None)
kp2,des2=orb.detectAndCompute(cereals,None)
```

```python
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
matches=bf.match(des1,des2)
```

```python
matches=sorted(matches,key=lambda x:x.distance)
```

```python
reeses_matches=cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:20],None,flags=2)  # matches upto 20
```

```python
display(reeses_matches)
```

```python
reeses_matches=cv2.drawMatches(reeses,kp1,cereals,kp2,matches,None,flags=2)   # all matches
```

```python
display(reeses_matches)
```
